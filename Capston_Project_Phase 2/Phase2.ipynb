```python
# In the dataset, a 'TOTAL' row is present for each STATE/UT and DISTRICT.
# This row aggregates values across all other variables, potentially inuencing query outcomes due to its cumulative nature.
# To ensure simplicity and accuracy in analysis, it is advisable to remove rows containing 'TOTAL' entries.

IPC_2001_2012 = IPC_2001_2012[IPC_2001_2012['DISTRICT'] != 'TOTAL'].reset_index(drop = True)
```

```python
# Renaming DELHI UT

IPC_2001_2012['STATE/UT'] = IPC_2001_2012['STATE/UT'].replace('DELHI UT', 'DELHI')
```

```python
# Adding up all the crimes and storing it into new 'TOTAL IPC CRIMES' column

IPC_2001_2012['TOTAL IPC CRIMES'] = IPC_2001_2012.iloc[:, 2:-1].sum(axis = 1)

IPC_2001_2012
```

We have NaNs in Total crimes against STs,Total crime against children columns


```python
# Filling NaNs with Mean

data_df['Total crimes against STs']=data_df['Total crimes against STs'].fillna(data_df['Total crimes against STs'].mean())

data_df['Total crime against children']=data_df['Total crime against children'].fillna(data_df['Total crime against children'].mean())

data_df['total crime against women']=data_df['total crime against women'].fillna(data_df['total crime against women'].mean())
```

## We have cleaned our data will move to Visulization

## Chart 1

### State/UT VS Literacy Rate

```python
plt.figure(figsize=(15,6))

ax = sns.barplot(x='State/UT', y='Litracy_rate',data=data_df)

# Adding data labels
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 10),
                textcoords = 'offset points')

plt.title('State/UT VS Literacy Rate')
plt.xlabel('State/UT')
plt.ylabel('Literacy Rate')
plt.xticks(rotation='vertical')
plt.show()
```

**Insights from above chart,**
- The x-axis represents the different states/union territories, and the y-axis represents the literacy rates.
- Kerala has the Highest Litracy_rate and Bihar has the lowest Literacy rate while bihar is lowest one
