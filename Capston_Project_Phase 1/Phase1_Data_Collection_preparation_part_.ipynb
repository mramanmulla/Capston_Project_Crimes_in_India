```python
# By using linear interpolation method, will calculate population and literacy rate  for between years that is 2002 to 2011

# Firstly for Population


def linear_interpolation(start_year, end_year, state_data, population_col_prefix='POPULATION_'):
    # Calculate the annual change in population
    start_col = f'{population_col_prefix}{start_year}'
    end_col = f'{population_col_prefix}{end_year}'
    state_data['Annual Change'] = (state_data[end_col] - state_data[start_col]) / (end_year - start_year)

    # Estimate population for intermediate years
    intermediate_years = range(start_year + 1, end_year)
    for year in intermediate_years:
        column_name = f'{population_col_prefix}{year}'
        state_data[column_name] = (state_data[start_col] + state_data['Annual Change'] * (year - start_year)).round()

    # Drop the 'Annual Change' column
    state_data.drop(columns=['Annual Change'], axis=1, inplace=True)

# Example usage
start_year = 2001
end_year = 2011

linear_interpolation(start_year, end_year, state_data, population_col_prefix='POPULATION_')
```

```python
import pandas as pd

def linear_interpolation(start_year, end_year, state_data, literacy_col_prefix='LITERACY_RATE_'):
    # Calculate the annual change in literacy rate
    start_col = f'{literacy_col_prefix}{start_year}'
    end_col = f'{literacy_col_prefix}{end_year}'
    state_data['Annual Change'] = (state_data[end_col] - state_data[start_col]) / (end_year - start_year)

    # Estimate literacy rate for intermediate years
    intermediate_years = range(start_year + 1, end_year)
    for year in intermediate_years:
        column_name = f'{literacy_col_prefix}{year}'
        state_data[column_name] = (state_data[start_col] + state_data['Annual Change'] * (year - start_year)).round(2)  # Assuming literacy rate is a percentage (round to 2 decimal places)

    # Drop the 'Annual Change' column
    state_data.drop(columns=['Annual Change'], axis=1, inplace=True)

# Example usage
start_year = 2001
end_year = 2011

linear_interpolation(start_year, end_year, state_data, literacy_col_prefix='LITERACY_RATE_')
```
